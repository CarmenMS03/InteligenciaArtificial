**_Ejercicio 4_**. Ejecuta el siguiente código entendiendo su funcionamiento y mejoralo para que vuelva a pedir la entrada cuando haya un fallo y para que calcule estadísticas adicionales como el rango, la mediana y la moda.
Para calcular la _moda_ investiga cómo funciona la función Counter (hay que importarla de collections)

```python
import math

def calcular_estadisticas(entrada):
    try:
        numeros = [float(x) for x in entrada.split(',')]
    except ValueError:
        print("Error: la entrada debe ser una lista de números separados por comas")
        return
    
    if not numeros:
        print("Error: No hay números.")
        return
    
    promedio = sum(numeros) / len(numeros)
    maximo = max(numeros)
    minimo = min(numeros)
    
    # Calcular la desviación estándar
    suma_cuadrados = sum((x - promedio) ** 2 for x in numeros)
    desviacion_estandar = math.sqrt(suma_cuadrados / len(numeros))
    
    print(f"Promedio: {promedio}")
    print(f"Número más grande: {maximo}")
    print(f"Número más pequeño: {minimo}")
    print(f"Desviación estándar: {desviacion_estandar}")
```

**_Ejercicio 6_**. Cálculo de áreas y volúmenes. 
Observa el uso de funciones en el siguiente código. Añade una opción de salir del menú y añade un bucle para repetir el menú principal hasta que el usuario decida salir.
Si el usuario ingresa un número que no corresponde a ninguna opción, mostrar un mensaje de error adecuado y permitirle volver a elegir opción.

```python
import math

def menu():
    print("Selecciona una figura geométrica:")
    print("1. Círculo")
    print("2. Esfera")
    print("3. Triángulo rectángulo")
    return int(input("Ingresa el número de la opción: "))

def calcular_area_circulo():
    radio = float(input("Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"El área del círculo es: {area:.2f}")

def calcular_volumen_esfera():
    radio = float(input("Ingresa el radio de la esfera: "))
    volumen = (4 / 3) * math.pi * radio ** 3
    print(f"El volumen de la esfera es: {volumen:.2f}")

def calcular_triangulo_rectangulo():
    base = float(input("Ingresa la base del triángulo: "))
    altura = float(input("Ingresa la altura del triángulo: "))
    area = 0.5 * base * altura
    hipotenusa = math.sqrt(base ** 2 + altura ** 2)
    print(f"El área del triángulo es: {area:.2f}")
    print(f"La hipotenusa del triángulo es: {hipotenusa:.2f}")

# Programa principal
opcion = menu()

if opcion == 1:
    calcular_area_circulo()
elif opcion == 2:
    calcular_volumen_esfera()
elif opcion == 3:
    calcular_triangulo_rectangulo()
else:
    print("Opción no válida. Intenta de nuevo.")
```
